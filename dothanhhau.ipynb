```python
df.drop(columns=['Dấu thời gian', 'Ngành học.1'], inplace=True)
df
```

```python
questions = df.drop(columns=['Tên Sinh viên', 'Ngành học', 'Giới tính'])
```

```python
questions['1. Tôi có tính tự lập'].unique()
answers = ['Chưa bao giờ đúng', 'Chỉ đúng 1 vài trường hợp', 'Chỉ đúng 1 nửa', 'Gần như là đúng', 'Hoàn toàn đúng']

questions = questions.map(lambda x: answers.index(x))
questions
```

```python
import seaborn  as sns
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize = (18, 18))
sns.heatmap(questions.corr(), annot=True, linewidths=.5, fmt='.1f', ax=ax)
```

```python

# Tìm các cột tương quan với ngưỡng x (arr: ma trận tương quan, x: giá trị tương quan)
def timCacCotTuongQuanVoiNguongX(arr, x):
  res = []
  n = len(arr.columns)
  for i in range(0, n):
    col = arr.columns[i]
    tmp = {col}
    for j in range(i + 1, n):
      row = arr.columns[j]
      if abs(arr.loc[row, col]) > x:
        tmp.add(row)
    if len(tmp) == 1: continue
    res.append(tmp)
  return res

# Tính ma trận tương quan
arr = questions.corr()

# Thiết lập ngưỡng tương quan, ví dụ như 0.5
x = 0.5

correlated_pairs = timCacCotTuongQuanVoiNguongX(arr, x)

print(len(correlated_pairs))
for x in correlated_pairs:  
  for y in x:
    print(y)
  print()
```

```python

pre = -1
cur = len(correlated_pairs)

while cur != pre:
  pre = cur
  visited = [False] * cur 
  res = []
  for i in range(cur):
    if visited[i]: continue
    visited[i] = True
    tmp = correlated_pairs[i]
    for j in range(i + 1, cur):
      uu = tmp & correlated_pairs[j]
      if not visited[j] and len(uu) > 0:
        tmp.update(correlated_pairs[j])
        visited[j] = True
    res.append(tmp)
  
  correlated_pairs = res
  cur = len(res)
    

for x in correlated_pairs:
  for y in x:
    print(y)
  print()
  
```

```python
#phương pháp chọn k thuộc tính tốt nhất
from sklearn.feature_selection import SelectKBest, chi2

X = questions
y = df['Ngành học']
# find best scored 35 features
k = 35  # Ví dụ, chọn 35 đặc trưng tốt nhất
selector = SelectKBest(chi2, k=k)
X_new = selector.fit_transform(X, y)

# In ra các đặc trưng đã chọn
selected_features = X.columns[selector.get_support()]
selected_features
```
